```python
import matplotlib.pyplot as plt

# Data for Median Household Income
years = [2023, 2024, 2025, 2026, 2027, 2028]
income = [72.7, 76.3, 79.8, 83.1, 86.1, 89.0]

# Median Household Income bar chart
plt.figure(figsize=(10, 6))
plt.bar(years, income, color='darkblue')
plt.title('Median Household Income in Odessa (2023-2028)')
plt.xlabel('Year')
plt.ylabel('Income ($ thousands)')
plt.xticks(years)
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.text(0.5, -0.1, "Source: Moody's Analytics", ha='center', va='top', transform=plt.gca().transAxes)
plt.show()
```
